```python
# the 'Açıklama' field to categorize transactions
def categorize_negative_transaction(description):
    description = description.lower()
    if "sabancı unıversıtesı" == description :
        return "Cafeteria food"
    elif "market" in description or "a101" in description or "migros" in description or "harun" in description or "baran" in description or ("sok" in description and "sabancı" in description) or "gıda" in description or "7036" in description:
        return "Market – grocery"
    elif "udemy" in description or "education" in description or "yds google" in description:
        return "Education"
    elif "save" in description or "investment" in description:
        return "Savings"
    elif "entertainment" in description or "game" in description or "fikret" in description or "alp dereli" in description or "mecra" in description or "amznprimetr" in description or "dr.com" in description or "asude" in description or "kitap.com" in description:
        return "Entertainment"
    elif "restaurant" in description or "restoran" in description or "yemek" in description or "iyzıco/matikotomat.c" == description or "kucuk ev" in description or "otomat" in description or "ırmak" in description:
        return "Dining out - food"
    elif "coffee" in description or "starbucks" in description or "cafem" in description or "bufe" in description or "kafe" in description or "cafe" in description or "simit" in description:
        return "Dining out - coffee"
    elif "fashion" in description or "clothing" in description or "shop" in description or "trendyol" in description or "hm.com" in description or "zara" in description or "pull bear" in description  or "bershka" in description:
        return "Fashion"
    elif "temiz" in description or "taksi" in description or "obilet" in description or "gur-sel" in description or "belbim" in description or "asıs" in description or "taksı" in description:
        return "Transportation"
    elif "para çekme" in description:
        return "ATM Use"
    else:
        return "Others"
```

```python
def categorize_incoming_transaction(description):
    description = description.lower()
    if "sabancı universitesi" in description or "kısrakdere kömürleri maden" in description:
        return "Grant"
    else:
        return "Allowance"
```

```python
# non-standard Turkish characters
def normalize_turkish_characters(text):
    if not isinstance(text, str):
        return ""
    replacements = {
        'ý': 'ı',
        'þ': 'ş',
        'ð': 'ğ',
        'Ý': 'İ',
        'Þ': 'Ş',
        'Ð': 'Ğ'
    }
    for old_char, new_char in replacements.items():
        text = text.replace(old_char, new_char)
    return text

# Apply normalization to the 'Açıklama' column
data["Açıklama"] = data["Açıklama"].apply(normalize_turkish_characters)

# data
```

```python
import os


# folder path 
folder_path = r'C:\Users\Lenovo\Desktop\DSA210_raw_Data'

# list of files (terms of school) to process
file_names = ["hist_5.csv", "hist_4.csv", "hist_3.csv", "hist_2.csv", "hist_1.csv"]

all_data = []

for file_name in file_names:
    file_path = os.path.join(folder_path, file_name)
    data_rows = []
    with open(file_path, 'r', encoding='ISO-8859-1') as file:
        for line in file:
            line_data = line.strip().split(';')
            data_rows.append(line_data)

    headers = data_rows[6]  # Adjust the header index as needed
    data = pd.DataFrame(data_rows[7:-5], columns=headers) # only the data points turns to a df

    # rename columns
    data.columns = ["Tarih", "Tutar", "Bakiye", "Açıklama", "Sub-category"]

    # normalize Turkish characters
    # fillna("") is to make empty rows for particular cols filled up with strings ""
    data["Açıklama"] = data["Açıklama"].fillna("").apply(normalize_turkish_characters)

    # 'Tutar' is numeric
    data["Tutar"] = (
        data["Tutar"]
        .str.replace('.', '', regex=False)  # remove thousands separator
        .str.replace(',', '.', regex=False)  # convert decimal separator
        .str.replace(' TL', '', regex=False)  # remove the currency symbol
        .astype(float)
    )
    
    # categorize transactions
    # categorization takes places in two for incomes and spendings
    data["Sub-category"] = data.apply(
        lambda row: categorize_negative_transaction(row["Açıklama"])
        if row["Tutar"] < 0
        else categorize_incoming_transaction(row["Açıklama"]),
        axis=1
    )

    all_data.append(data)

# all processed data for 5 terms
combined_data = pd.concat(all_data, ignore_index=True)

output_file_path = r'C:\Users\Lenovo\Desktop\DSA210_raw_Data\combined_categorized_transactions.csv'
combined_data.to_csv(output_file_path, index=False, encoding='utf-8')

print("All files processed and combined data saved to:", output_file_path)
```
